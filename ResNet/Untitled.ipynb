```python
class ResBlk(keras.layers.Layer):
    def __init__(self, out_chan, strides = 1, downsample = None, **kwargs):
        super().__init__()
        
        self.conv1 = keras.layers.Conv2D(out_chan, padding='same', kernel_size=3, strides=strides, use_bias = False)
        self.bn1 = keras.layers.BatchNormalization()
        self.relu1 = keras.layers.ReLU()
        
        self.conv2 = keras.layers.Conv2D(out_chan, padding='same', kernel_size=3, strides=strides, use_bias = False)
        self.bn2 = keras.layers.BatchNormalization()
        
        self.downsample = downsample
        
        self.relu2 = keras.layers.ReLU()
        
        print("init good")

    
    def call(self, X, training = False, **kwargs):
        y = X
        print("1")
        y = self.conv1(y)
        print("2")
        y = self.bn1(y, training)
        y = self.relu1(y)
        y = self.conv2(y)
        y = self.bn2(y, training)
        print("3")
        h_x = X
        print("4")
        if self.downsample != None:
            h_x = self.downsample(h_x)
        print("4")
        y += h_x
        y = self.relu2(y)
        return y
```

```python
blk = ResBlk(3)
```
